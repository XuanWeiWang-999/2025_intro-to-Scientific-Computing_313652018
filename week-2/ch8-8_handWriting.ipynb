# Solution:

We propose

$$
Hf(x) = \sum_{j=0}^n a_j (x-x_0)^j.
$$

Taking derivatives, we get

$$
(Hf)^{(k)}(x) = \sum_{j=k}^n a_j \frac{j!}{(j-k)!} (x-x_0)^{j-k}.
$$

Evaluating at $x=x_0$:

$$
(Hf)^{(k)}(x_0) = k! \, a_k.
$$

From the interpolation condition

$$
(Hf)^{(k)}(x_0) = f^{(k)}(x_0),
$$

we get

$$
a_k = \frac{f^{(k)}(x_0)}{k!}.
$$

Thus the polynomial is

$$
Hf(x) = \sum_{j=0}^n \frac{f^{(j)}(x_0)}{j!} (x-x_0)^j.
$$
$\blacksquare$
