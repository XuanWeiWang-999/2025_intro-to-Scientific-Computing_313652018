## Claim:

From Exercise 5 we have
$$
\omega_{n+1}(x) = (x+Nh)(x+(N-1)h)\cdots(x+h)x(x-h)\cdots(x-Nh),
$$
where $N=n/2$, $h=2/n$.



Consider the ratio
$$
\begin{aligned}
\frac{\omega_{n+1}(x+h)}{\omega_{n+1}(x)}&= \prod_{k=-N}^N \frac{x+h-kh}{x-kh}\\
&= \prod_{k=-N}^N \frac{x-(k-1)h}{x-kh}.
\end{aligned}
$$

This telescopes to
$$
\frac{\omega_{n+1}(x+h)}{\omega_{n+1}(x)}
= \frac{x+Nh}{x-Nh}.
$$

For $x \in (0, x_{n-1})$ (and $x$ not an interpolation node), we have $x+Nh > 0$, and $x-Nh < 0$.

Hence
$$
\left|\frac{\omega_{n+1}(x+h)}{\omega_{n+1}(x)}\right|
= \frac{x+Nh}{Nh-x} > 1.
$$


Thus $|\omega_{n+1}(x)|$ strictly increases as $x$ moves to the right.  
Therefore, the maximum value is achieved in the last interval:
$$
x \in (x_{n-1},x_n).
$$

Since $|\omega_{n+1}(x)|$ is even, the same maximum occurs also in $(-x_n,-x_{n-1})$.


Hence, $|\omega_{n+1}(x)|$ attains its maximum for $x \in (x_{n-1}, x_n)$.  
$\blacksquare$
