## Proof

Note that $\omega_{n+1}=\prod_{j=0}^{n}(x-x_j)$.  
Let $N=\frac{n}{2}$, then the nods are $$x_j=-1+jh,\ \ \text{for }j=0,...,n, h=\frac{2}{n}.$$
Hence, $$\omega_{n+1}(x) = (x+Nh)(x+(N-1)h)\cdots(x+h)x(x-h)\cdots(x-Nh).$$

We take $x = rh, \quad N-1 < r < N \quad (\text{since } x \in (x_{n-1},x_n)).$  
Then 
$$
\begin{aligned}
\omega_{n+1}(rh)&=\prod_{k=-N}^N (rh-kh)\\
&=h^{n+1}\prod_{k=-N}^N (r-k).
\end{aligned}
$$

Notice that $$(r-N)(r-(N-1))=\frac{1}{h^2}(x-x_n)(x-x_{n-1}).$$  
Then $$\omega_{n+1}(rh) 
= h^{\,n+1}(r-N)(r-(N-1)) \prod_{k=-N}^{N-2}(r-k).$$
We rewite it with $x$, then we get $$\omega_{n+1}(x) 
= h^{\,n-1}(x-x_{n-1})(x-x_n)\,\prod_{k=-N}^{N-2}(r-k).$$

We consider $P(r) = \prod_{k=-N}^{N-2}(r-k)$.  
Then there are $(n-1)$-factors, for $r\in ( N-1, N )$. Hence every fector is positive, and each of the value are similar to $(1,2,\dots,n-1)$.  
Hence we have the inequality:$$(n-1)! \leq P(r) \leq n!.$$

Then we have $$|\omega_{n+1}(x)| 
= h^{\,n-1}|(x-x_{n-1})(x-x_n)|\,|P(r)|.$$

Hence, $$(n-1)!h^{\,n-1}|(x-x_{n-1})(x-x_n)| 
\;\le\; |\omega_{n+1}(x)| 
\;\le\; n!h^{\,n-1}|(x-x_{n-1})(x-x_n)|.$$
$\blacksquare$
