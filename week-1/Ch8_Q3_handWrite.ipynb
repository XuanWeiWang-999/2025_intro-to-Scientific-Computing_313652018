# Ch8 #3



## Proof
Since $W_{n+1}(x)=\displaystyle\prod_{i=0}^{n}(x-x_i)=(x-x_0)(x-x_1)\cdots(x-x_n)$, 
we have 
$$
\begin{aligned}
W'_{n+1}(x) &=\frac{d}{dx}W_{n+1}(x) \\
            &=(x-x_1)(x-x_2)\cdots(x-x_n)+(x-x_0)[(x-x_2)(x-x_3)\cdots(x-x_n)+(x-x_1)[\cdots]] \\
            &=(x-x_1)(x-x_2)\cdots(x-x_n)+(x-x_0)(x-x_2)\cdots(x-x_n)+\cdots+(x-x_1)(x-x_2)\cdots(x-x_{n-1}) \\
            &=\displaystyle\prod_{\substack{j=0 \\j\neq i}}^{n}(x_i-x_j)
\end{aligned}$$

Then since $$\displaystyle\prod_{n}(x)=\displaystyle\sum_{i=0}^{n}y_il_i(x),$$ and 
            $$\displaystyle l_i(x)=\frac{\displaystyle\prod _{j\neq i}(x-x_j)}{\displaystyle\prod _{j\neq i}(x_i-x_j)},$$ 
    we thus checked formula (8.5).
