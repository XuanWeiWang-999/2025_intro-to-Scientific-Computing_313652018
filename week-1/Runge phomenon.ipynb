# Runge pheomenon

## 作業範例
$$f(x)=\frac{1}{1+x^2},\ \ -5\leq x\leq 5$$


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def runge_function(x):
    return 1 / (1 + x**2)

x_plot = np.linspace(-5, 5, 1000)
y_plot = runge_function(x_plot)

degrees = [5, 10, 15]
plt.figure(figsize=(10, 6))

for n in degrees:

    x_nodes = np.linspace(-5, 5, n+1)
    y_nodes = runge_function(x_nodes)

    # 插值多項式
    poly_coeffs = np.polyfit(x_nodes, y_nodes, n)
    y_interp = np.polyval(poly_coeffs, x_plot)

    plt.plot(x_plot, y_interp, label=f"n={n} interpolation polynomial")
    plt.scatter(x_nodes, y_nodes, marker="o")

# 真實函數
plt.plot(x_plot, y_plot, "k--", linewidth=2, label="f(x) = 1/(1+x^2)")

plt.title("Runge phenomenon: f(x)=1/(1+x^2), for range in [-5,5]")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```

## 作業結論
虛線表示最原始的$f(x)=\frac{1}{1+x^2}$函數，藍、橘、綠表示當平均間格取1000點時分別做5、10、15次插植多項式時所模擬出來的函數曲線。可以看見在中間高峰都模擬得很好，但一旦靠近邊界時就會有明顯震盪。
